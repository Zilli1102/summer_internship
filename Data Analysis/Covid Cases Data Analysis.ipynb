```python
# Filter and prepare data
asia_data = df[df['continent'] == 'Asia'].copy()
asia_data['date'] = pd.to_datetime(asia_data['date'])

# Group by week and sum new_cases and total_deaths
weekly_data = asia_data.groupby(pd.Grouper(key='date', freq='W'))[['new_cases', 'total_deaths']].sum().reset_index()

# Fill NaNs with 0
weekly_data.fillna(0, inplace=True)

# Plot
x = np.arange(len(weekly_data['date']))  # positions
width = 0.4  # width of each bar

plt.figure(figsize=(20, 8))
plt.bar(x - width/2, weekly_data['new_cases'], width=width, label='New Cases', color='skyblue')
plt.bar(x + width/2, weekly_data['total_deaths'], width=width, label='Total Deaths', color='salmon')

# Formatting
plt.xlabel('Week', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Weekly COVID-19 New Cases vs Total Deaths in Asia', fontsize=18)
plt.xticks(x[::4], weekly_data['date'].dt.strftime('%b %d')[::4], rotation=45, fontsize=10)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Filter and process Asia-specific data
asia_df = df[df['continent'] == 'Asia'].copy()
asia_df['date'] = pd.to_datetime(asia_df['date'])

# Group by date and sum up values
asia_grouped = asia_df.groupby('date')[['new_cases', 'total_deaths']].sum().reset_index()
asia_grouped.fillna(0, inplace=True)

# Plotting
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date", fontsize=14)
plt.ylabel("total_deaths", fontsize=14)
plt.title("COVID-19 Trend in Asia (New Cases vs Total Deaths)", fontsize=18)

# Plot lines
plt.plot(asia_grouped['date'], asia_grouped['new_cases'], label='New Cases', color='red')

# Optional axis limits (adjust as needed)
# plt.ylim(0, 350000)

# Add legend and grid
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Prepare and group the data
df['date'] = pd.to_datetime(df['date'])
grouped_df = df.groupby(['date', 'continent'])['total_deaths'].sum().reset_index()
reshaped_df = grouped_df.pivot(index='date', columns='continent', values='total_deaths')
reshaped_df.fillna(0, inplace=True)  # Replace NaNs with 0

# Plot
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Total Deaths", fontsize=14)
plt.title("Total COVID-19 Deaths Over Time by Continent", fontsize=18)
plt.ylim(0, reshaped_df.max().max() * 1.1)  # Adjust y-axis based on max

for column in reshaped_df.columns:
    plt.plot(reshaped_df.index, reshaped_df[column], label=column, linewidth=3)

plt.legend(fontsize=16)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show(block=False)
```

```python
# Ensure datetime format
df['date'] = pd.to_datetime(df['date'])

# Group and pivot data
grouped_df = df.groupby(['date', 'continent'])['total_deaths'].sum().reset_index()
reshaped_df = grouped_df.pivot(index='date', columns='continent', values='total_deaths')
reshaped_df.fillna(0, inplace=True)

# Apply 6-day rolling average
rolling_df = reshaped_df.rolling(window=6).mean()

# Plot
plt.figure(figsize=(16, 10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Total Deaths (6-Day Rolling Average)", fontsize=14)
plt.title("Rolling Average of COVID-19 Deaths by Continent", fontsize=18)
plt.ylim(0, rolling_df.max().max() * 1.1)

for column in rolling_df.columns:
    plt.plot(rolling_df.index, rolling_df[column], label=column, linewidth=3)

plt.legend(fontsize=16)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
