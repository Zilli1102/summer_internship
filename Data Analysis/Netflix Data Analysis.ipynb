```python
df['Score'] = pd.to_numeric(df['Score'], errors='coerce')  # convert to float, invalid parsing becomes NaN

result = df.groupby('[Country or]')['Score'].mean()
print(result)
```

```python
df['[GDP per capita]'] = pd.to_numeric(df['[GDP per capita]'], errors='coerce')
```

```python
df['GDP_bin'] = pd.cut(df['[GDP per capita]'], bins=5)
reshaped_df = df.pivot_table(
    index='[Country or]',
    columns='GDP_bin',
    values='Score',
    aggfunc='mean',
    observed=False
)
```

```python
import plotly.graph_objects as go

# Reset index to get 'Country' as a column
interactive_df = reshaped_df.reset_index()
interactive_df.rename(columns={interactive_df.columns[0]: 'Country'}, inplace=True)

fig = go.Figure()

# Add a line trace for each GDP bin column
for col in reshaped_df.columns:
    fig.add_trace(go.Scatter(
        x=interactive_df['Country'],
        y=interactive_df[col],
        mode='lines+markers',
        name=str(col),
        line=dict(width=3),
        marker=dict(size=6)
    ))

fig.update_layout(
    title="GDP Bin Distribution by Country",
    xaxis_title="Country",
    yaxis_title="No of Entries",
    yaxis=dict(range=[0, reshaped_df.values.max() + 1]),
    xaxis_tickangle=90,
    xaxis_tickfont=dict(size=14),
    yaxis_tickfont=dict(size=14),
    legend_title_text="GDP Bins",
    legend=dict(font=dict(size=12), title_font=dict(size=14)),
    margin=dict(l=60, r=20, t=80, b=150),
    height=600,
    hovermode='x unified'  # shows all lines' values at a vertical hover line
)

fig.show()
```

```python
import plotly.graph_objects as go

# Compute rolling mean with window=6 and min_periods=1
rolling_df = reshaped_df.rolling(window=6, min_periods=1).mean()

# Reset index for Plotly
interactive_rolling_df = rolling_df.reset_index()
interactive_rolling_df.rename(columns={interactive_rolling_df.columns[0]: 'Country'}, inplace=True)

fig = go.Figure()

# Add trace for each GDP bin column
for col in rolling_df.columns:
    fig.add_trace(go.Scatter(
        x=interactive_rolling_df['Country'],
        y=interactive_rolling_df[col],
        mode='lines+markers',
        name=str(col),
        line=dict(width=3),
        marker=dict(size=6)
    ))

fig.update_layout(
    title="Rolling Mean (Window=6) of GDP Bin Distribution by Country",
    xaxis_title="Country",
    yaxis_title="Rolling Mean of Entries",
    yaxis=dict(range=[0, rolling_df.max().max() + 1]),
    xaxis_tickangle=90,
    xaxis_tickfont=dict(size=14),
    yaxis_tickfont=dict(size=14),
    legend_title_text="GDP Bins",
    legend=dict(font=dict(size=12), title_font=dict(size=14)),
    margin=dict(l=60, r=20, t=80, b=150),
    height=600,
    hovermode='x unified'
)

fig.show()
```

```python
import plotly.graph_objects as go

# Select top 20 countries by total sum across columns
top_countries = reshaped_df.sum(axis=1).sort_values(ascending=False).head(20).index

# Subset and apply rolling mean (window=6, min_periods=1)
subset_rolling_df = reshaped_df.loc[top_countries].rolling(window=6, min_periods=1).mean()

# Reset index for Plotly and rename index column
interactive_subset_df = subset_rolling_df.reset_index()
interactive_subset_df.rename(columns={interactive_subset_df.columns[0]: 'Country'}, inplace=True)

fig = go.Figure()

# Add a line for each GDP bin column
for col in subset_rolling_df.columns:
    fig.add_trace(go.Scatter(
        x=interactive_subset_df['Country'],
        y=interactive_subset_df[col],
        mode='lines+markers',
        name=str(col),
        line=dict(width=3),
        marker=dict(size=6)
    ))

fig.update_layout(
    title="Rolling Mean (Window=6) for Top 20 Countries",
    xaxis_title="Country",
    yaxis_title="Rolling Mean",
    xaxis_tickangle=90,
    xaxis_tickfont=dict(size=14),
    yaxis_tickfont=dict(size=14),
    legend_title_text="GDP Bins",
    legend=dict(font=dict(size=12), title_font=dict(size=14)),
    margin=dict(l=60, r=20, t=80, b=150),
    height=600,
    hovermode='x unified'
)

fig.show()
```

```python
import matplotlib.pyplot as plt

top_countries = reshaped_df.sum(axis=1).sort_values(ascending=False).head(10).index
subset_df = reshaped_df.loc[top_countries]

num_bins = len(reshaped_df.columns)
cols = 3
rows = (num_bins + cols - 1) // cols

fig, axes = plt.subplots(rows, cols, figsize=(18, rows * 4), sharey=True)
axes = axes.flatten()

for i, col in enumerate(reshaped_df.columns):
    ax = axes[i]
    for country in top_countries:
        ax.plot(subset_df.index, subset_df[col], label=country if i==0 else "", alpha=0.7)
    ax.set_title(f'GDP Bin {col}')
    ax.tick_params(axis='x', rotation=90)
    if i % cols == 0:
        ax.set_ylabel('No of Entries')
    if i >= len(reshaped_df.columns) - cols:
        ax.set_xlabel('Country')

# Hide empty subplots if any
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.legend(loc='upper left')
plt.show()
```
